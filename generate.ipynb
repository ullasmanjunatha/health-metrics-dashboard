```python
import numpy as np
import pandas as pd
import h5py

n_patients = 60000
np.random.seed(123)

ages = np.random.randint(18, 90, n_patients)
bmi = np.round(np.random.normal(25, 5, n_patients), 1)
cholesterol = np.random.choice(["normal", "high", "very high"], size=n_patients, p=[0.6, 0.3, 0.1])
glucose = np.random.choice(["normal", "high", "very high"], size=n_patients, p=[0.5, 0.3, 0.2])
activity = np.random.randint(0, 10, n_patients)
disease = np.random.choice([0, 1], size=n_patients, p=[0.85, 0.15])

with h5py.File("health_metrics.h5", "w") as f:
    f.create_dataset("age", data=ages)
    f.create_dataset("bmi", data=bmi)
    f.create_dataset("cholesterol", data=np.array(cholesterol.astype('S')))
    f.create_dataset("glucose", data=np.array(glucose.astype('S')))
    f.create_dataset("activity_level", data=activity)
    f.create_dataset("has_disease", data=disease)
```
